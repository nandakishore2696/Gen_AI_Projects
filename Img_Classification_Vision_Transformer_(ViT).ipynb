# import required Libs

```python
from transformers import ViTFeatureExtractor, ViTForImageClassification
from PIL import Image
import requests
```

# Download Image from the given URL

```python
#url = 'http://images.cocodataset.org/val2017/000000039769.jpg'

url = "https://images.moneycontrol.com/static-mcnews/2024/03/market_stocks_sensex_Nifty-6.jpg?impolicy=website&width=1600&height=900"
image = Image.open(requests.get(url, stream=True).raw)
```

# Display The downloaded image

```python
#from IPython.display import Image
display(image)
```

# Download Pre-trained Model

```python
feature_extractor = ViTFeatureExtractor.from_pretrained('google/vit-base-patch16-224')
model = ViTForImageClassification.from_pretrained('google/vit-base-patch16-224')
```

# Extract Features

```python
inputs = feature_extractor(images=image, return_tensors="pt")
```

```python
inputs
```

# Model Inference

```python
outputs = model(**inputs)
# or use outputs = model(inputs['pixel_values'])
```

```python
outputs.keys()
```

```python
logits = outputs.logits
```

# model predicts one of the 1000 ImageNet classes

```python
predicted_class_idx = logits.argmax().item()
print("Predicted class:", model.config.id2label[predicted_class_idx])
```
